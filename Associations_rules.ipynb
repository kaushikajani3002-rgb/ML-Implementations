```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# using an in-built dataset: Grocery transaction
dataset = [['milk', 'bread', 'butter'],
           ['bread', 'butter'],
           ['milk', 'bread', 'butter', 'cheese'],
           ['milk', 'bread'],
           ['butter', 'cheese'],
           ['bread', 'butter', 'cheese']]

```

```python
# step 1: convert the dataset into a one-hot encoderr data-frame

te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_data,  columns=te.columns_)

```

```python
# step 2 apply  the apriori algorithm to find frequet itemsets
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

```

```python
# stpe 3 generate association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)

#output the result 

print("Frequent Itemsets")
print(frequent_itemsets)
print("\n Association Rules:")
print(rules)
```
